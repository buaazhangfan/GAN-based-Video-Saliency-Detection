```python
import matplotlib.pyplot as plt


xlable_list = ['AUC', 'NSS', 'KL', 'EMD', 'CC']
GBVS_value = [0.60, 0.48, 1.10, 1.88, 0.30]
Xu_value = [0.64, 0.54, 1.10, 1.95, 0.33]
Itti_value = [0.50, 0.32, 17.02, 2.27, 0.22]
PQFT_value = [0.50, 0.21, 19.76, 3.28, 0.15]
Hou_value = [0.53, 0.54, 6.98, 2.56, 0.28]
Judd_value = [0.71, 0.81, 0.98, 1.80, 0.38]

x = list(range(len(xlable_list)))
plt.figure(figsize=(12,7))  
total_width, n = 0.8, 6
width = total_width / n

plt.bar(x, GBVS_value, width = width, label = 'GBVS', fc = 'royalblue')
for i in range(len(x)):
	x[i] = x[i] + width
plt.bar(x, Xu_value, width = width, label = 'Xu', fc = 'limegreen')
for i in range(len(x)):
	x[i] = x[i] + width
plt.bar(x, Itti_value, width = width, label = 'Itti', tick_label = xlable_list, fc = 'coral')
for i in range(len(x)):
	x[i] = x[i] + width
plt.bar(x, PQFT_value, width = width, label = 'PQFT', fc = 'lightseagreen')
for i in range(len(x)):
	x[i] = x[i] + width
plt.bar(x, Hou_value, width = width, label = 'Hou', fc = 'violet')
for i in range(len(x)):
	x[i] = x[i] + width
plt.bar(x, Judd_value, width = width, label = 'Judd', fc = 'orange')

plt.legend()  
plt.xlabel("Benchmark")  
plt.ylabel("Score")  
plt.title("Performance score for all models(Traffic)", bbox={'facecolor':'0.8', 'pad':5})
plt.grid(True)  
plt.savefig("/Users/apple/Desktop/plot/Traffic.jpg")
plt.show()  
```
